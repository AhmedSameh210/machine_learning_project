```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv('original.csv')
```

<font size='+20'> <b> Helper Function </font> </b>

```python
def Convert(entry) -> set:
    my_set = set()
    flag = False
    current_word = ""
    for j in range(1, len(entry)):
        currnet_char = entry[j]
        if(currnet_char == '\'' and flag):
            my_set.add(current_word)
            current_word = ""
            flag = False
        elif(currnet_char == '\''):
            flag = True
        elif j != len(entry) - 1  and currnet_char != ',':
            current_word += currnet_char
    return my_set
```

<font size='+20'> <b> Artist(s) Genres Column </font> </b>

```python
sub_data = pd.DataFrame({'Artist Names':data['Artist Names'],'Artist(s) Genres':data['Artist(s) Genres']})
dic = {}

for i in range(0,len(sub_data)):
    
    entry_of_artists = tuple(Convert(sub_data.iloc[i,0]))
    entry_of_artists_genres = Convert(sub_data.iloc[i,1])
    
    if dic.get(entry_of_artists) == None:
        dic.update({entry_of_artists : entry_of_artists_genres})
    else:
        dic[entry_of_artists] = entry_of_artists_genres.union(dic[entry_of_artists])
```

<font size='+20'> <b> Genres Count Column </font> </b>

```python
list_of_genres_count = []

for element in dic.values():
    list_of_genres_count.append(len(element))
```

<font size='+20'> <b> Weight Column </font> </b>

```python
list(dic.values())
```

```python
genres_freq = {}
weight_column = [0 for _ in range(0,len(data['Artist(s) Genres']))]
for entry in list(dic.values()):
    for gener in entry:
        if genres_freq.get(gener) == None:
            genres_freq.update({gener:1})
        else:
            genres_freq[gener] += 1

for i,entry in enumerate(list(dic.values())):
    for gener in entry:
        weight_column[i] += genres_freq[gener]
```

<font size='+20'> <b> Exporting the Csv </font> </b>

```python
with open('abdelrhman&nour.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Song', 'Artist(s) Genres','Genres count','Weight','total weight of genres'])  # Writing header
    for item1, item2, item3, item4, item5 in zip(list(data['Song']), list(dic.values()), list_of_genres_count, weight_column, [x * y for x, y in zip(list_of_genres_count, weight_column)]):
        writer.writerow([item1, item2, item3, item4, item5])
```
